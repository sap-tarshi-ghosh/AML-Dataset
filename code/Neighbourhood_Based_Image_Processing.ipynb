```python
from PIL import Image

image = Image.open('low-contrast-04.jpg').convert('L')

width, height = image.size

pixels = list(image.getdata())

#converting the image into a 2d list
img = []
for y in range(height):
  row = []
  for x in range(width):
    row.append(pixels[y * width + x])
  img.append(row)

```

```python
mean_img = []

for y in range(height):
  row = []
  for x in range(width):
    row.append(img[y][x])
  mean_img.append(row)
```

# Applying mean filter

```python
for y in range(height):
  for x in range(width):
    total = 0

    temp = [-1, 0, 1]
    for dy in temp:
      for dx in temp:
        total += img[y + dy][x + dx]

    mean_img[y][x] = int(total/9)
```

```python
for y in range(1, height -1):
  for x in range(1, width - 1):
    total = 0

    temp = [-1, 0, 1]
    for dy in temp:
      for dx in temp:
        total += img[y + dy][x + dx]

    mean_img[y][x] = int(total/9)
```

Visulising the picture

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title('original image')

plt.subplot(1, 2, 2)
plt.imshow(mean_img, cmap='gray', vmin=0, vmax=255)
plt.title('mean filtered image (3X3)')
plt.axis('off')

plt.show()
```
